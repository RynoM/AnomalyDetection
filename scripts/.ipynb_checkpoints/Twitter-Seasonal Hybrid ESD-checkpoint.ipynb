```r
library(AnomalyDetection)
```

```r
MyData <- read.csv(file="Anomaly detection/data/MV_fake-autocorr-test.csv", header=TRUE, sep=",")
```

```r
columnlist = list('X0', 'X1', 'X2', 'X3', 'X4', 'X5')
```

```r
for (var in columnlist){
    res = AnomalyDetectionTs(data.frame(MyData$X,MyData[[var]]), max_anoms=0.01, direction='both', plot=TRUE, longterm=FALSE, alpha=1.5)
    write.csv(res$anoms, file = paste("twitter anomalies/Twitter-anoms",unlist(strsplit(var,""))[2],".csv", sep=""))
}
```

```r
res = AnomalyDetectionTs(data.frame(MyData$X,MyData$X0), max_anoms=0.1, direction='both', plot=TRUE, longterm=FALSE, alpha=1.5)
res$plot
```
