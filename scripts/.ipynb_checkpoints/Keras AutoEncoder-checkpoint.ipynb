## Pre-processing functions

```python
def robust_normalize(data):
    ''' Create scaler and fit to values'''
    scaler = RobustScaler()
    return scaler.fit_transform(data)


def create_trainset(dataset, look_back=1):
    ''' Create RNN input data, scaled with RobustScaler,
    for each input window of size 'look_back+1'. '''
    dataX = []
    for i in range(look_back, len(dataset)):
        dataX.append(robust_normalize(dataset.iloc[i-look_back:i+1]))
    return np.array(dataX)


def robust_recreate(predictions, timesteps, func):
    ''' Make one prediction from list of predictions per point.
    Possible methods to combine are; mean, median, min, max and last. ''' 
    dfs = {}  #dict to store df's for timesteps
    for y in range(timesteps): 
        pred = []
        for i in range(len(predictions)):
            pred.append(predictions[i][y])
        dfs[y] = pd.DataFrame(pred, index=range(y,(len(predictions)+y))) #recenter predictions based on timestep index
    df_concat = pd.concat(dfs.values())
    if func == 'mean':
        return df_concat.groupby(df_concat.index).mean() #groupby all same indexes and perform function on them
    elif func == 'median':
        return df_concat.groupby(df_concat.index).median()
    elif func == 'max':
        df_max = pd.DataFrame(index=df_concat.index.unique(), columns=range(predictions.shape[2]))
        for ix in df_concat.index.unique():
            sub = df_concat[df_concat.index == ix]
            for element in sub:
                df_max.iloc[ix, element] = max(sub[element].values, key=abs)
        return df_max.astype(float)
    elif func == 'min':
        df_min = pd.DataFrame(index=df_concat.index.unique(), columns=range(predictions.shape[2]))
        for ix in df_concat.index.unique():
            sub = df_concat[df_concat.index == ix]
            for element in sub:
                df_min.iloc[ix, element] = min(sub[element].values, key=abs)
        return df_min.astype(float)
    elif func == 'last':
        return dfs[timesteps-1]


def seasonal_decomp(dataset, freq=24*7, remove_trend=True, removal='Normal'):
    ''' Decomposition based on statsmodels seasonal_demposition
    Only seasonal component is subtracted from original as trend is also
    considered anomalous in short timeframes. '''
    
    deseason = pd.DataFrame(columns=dataset.columns, index=dataset.index)
    for column in dataset:
        deq = decompose(dataset[column].values, period=freq)
        if remove_trend == True:
            if removal == 'Normal':
                deseason[column] = dataset[column] - deq.seasonal - deq.trend
            elif removal == 'Median':
                deseason[column] = dataset[column] - deq.seasonal - dataset[column].median()
        else:
            deseason[column] = dataset[column] - deq.seasonal 
    return deseason


def autocorr_lag1(signal):
    return np.corrcoef(signal[:-1], signal[1:])[0][1]
```

## Build AutoEncoder

```python
from keras import optimizers
from keras.layers import Input
from keras.models import Model
from keras.layers.recurrent import GRU
from keras.layers.core import RepeatVector

timesteps, input_dim = 24, 6

inputs = Input(shape=(timesteps, input_dim))
# hidden = GRU(20, return_sequences=True)(inputs)
hidden = GRU(12, return_sequences=False)(inputs)
encoded = RepeatVector(timesteps)(hidden)
hidden = GRU(12, return_sequences=True)(encoded)
# hidden = GRU(20, return_sequences=True)(hidden)
outputs = GRU(input_dim, return_sequences=True)(hidden)

model = Model(inputs=inputs, outputs=outputs)
rmsprop = optimizers.RMSprop(lr=0.012)
model.compile(loss='mean_squared_error', optimizer=rmsprop)
```

## Anomaly based on multivariate distance (Mahalanobis)

```python
def MV_gaussian(errors):
    d_list = []
    distances = pd.DataFrame(index=errors.index, columns=['distances'])
    mv_mean = np.mean(errors, axis=0)
    robust_cov = MinCovDet().fit(errors)
    for pred_error in errors.iterrows():
        d_list.extend(robust_cov.mahalanobis([pred_error[1] - mv_mean]))
    distances['distances'] = d_list
    return distances


def MV_anom_threshold(dists, threshold):
    anoms=[]
    for ix, err in enumerate(dists):
        if err > threshold:
            anoms.extend([ix])
    return anoms

def anomaly_kernal_MV(errors, UB):
    anomalies = []
    kernal = gaussian_kde(errors['distances'])
    for ix, point in errors.iterrows():
        integral = kernal.integrate_box_1d(point[0], np.inf)
        if integral < UB:
            anomalies.append(ix)
    return anomalies


#kernal density estimation deals with the skewedness of the distributions
#distributions are observed to be unimodal, kde works better on unimodal, multimodel tends to be oversmoothed. 
#scipy has automatic bandwith est.
def estimate_kernals(errors):
    kernals = {}
    for column in errors:
        kernals[column] = gaussian_kde(errors[column].astype(float))
    return kernals


def gridsearch(UBgrid, LBgrid, anoms, errors, kernal):
    grid = list(itertools.product(*[UBgrid, LBgrid]))
    perf = {}
    for (UB,LB) in grid:
        anomalies = anomaly_kernal(errors, kernal, UB, LB)
        perf[UB,LB] = F1_performance(anoms, anomalies, 672)[1]
    return max(perf, key=lambda key: perf[key]), perf[max(perf, key=lambda key: perf[key])]


def anomaly_kernal(errors, kernal, UB, LB):
    anomalies = []
    for ix, point in enumerate(errors):
        integral = kernal.integrate_box_1d(point, np.inf)
        if integral > UB or integral < LB:
            anomalies.append(ix)
    return anomalies


def gridsearch_MV(UBgrid, LBgrid, anoms, errors):
    grid = list(itertools.product(*[UBgrid, LBgrid]))
    perf = {}
    CM = {}
    TN = len(errors)-len(anoms)
    kernals = estimate_kernals(errors)
    for (UB,LB) in grid:
        anomalies = []
        for kernal in kernals:
            anomalies.extend(anomaly_kernal(errors[kernal].values, kernals[kernal], UB, LB))
        perf[UB,LB] = F1_performance(anoms, anomalies, TN)[1]
        CM[F1_performance(anoms, anomalies, TN)[1]] = F1_performance(anoms, anomalies, TN)[0]
    return max(perf, key=lambda key: perf[key]), \
            perf[max(perf, key=lambda key: perf[key])], CM[perf[max(perf, key=lambda key: perf[key])]]
    
    
def F1_performance(true_labels, pred_labels, length):
    ''' Calculate F1-score based on the true and predicted labels.
    Length of the original dataset is needed for FN. '''
    
    CM = pd.DataFrame({"pred_T": {"label_T": len(set(true_labels) & set(pred_labels)), 
                                    "label_F": len(set(pred_labels) - set(true_labels))},
                         "pred_F": {"label_T": len(set(true_labels) - set(pred_labels)),
                                    "label_F": len(set(range(0, length)) - set(true_labels)- set(pred_labels))}}) 
    
    recall = CM['pred_T']['label_T'] / (CM['pred_T']['label_T'] + CM['pred_F']['label_T'])
    precision = CM['pred_T']['label_T'] / (CM['pred_T']['label_T'] + CM['pred_T']['label_F'])
    print('recall: ', recall)
    print('precision: ', precision)
    F1 = 2 * (precision * recall) / (precision + recall)
    return CM, F1

    
def robust_rescale(original, predictions):
    for ix, column in enumerate(original):
        scaler = RobustScaler()
        scaler = scaler.fit(original[column].values.reshape(-1,1))
        predictions[column] = scaler.inverse_transform(predictions[ix].values.reshape(-1,1))
    return predictions[original.columns]
```

## Anomalies based on Multivariate error recreation

```python
def anomaly_kernal_new(errors, kernal, UB, LB):
    anomalies = []
    for ix, point in errors.iteritems():
        integral = kernal.integrate_box_1d(point, np.inf)
        if integral > UB or integral < LB:
            anomalies.append(ix)
    return anomalies


def F1_performance_new(true_labels, pred_labels, n_points, n_vars):
    
    TP, FP, FN, TN = 0, 0, 0, 0
    
    for x in range(n_vars):
        TP += len(set(true_labels[str(x)]) & set(pred_labels[pred_labels['anoms'] == x]['ix']))
        FP += len(set(pred_labels[pred_labels['anoms'] == x]['ix']) - set(true_labels[str(x)]))
        FN += len(set(true_labels[str(x)]) - set(pred_labels[pred_labels['anoms'] == x]['ix']))
        TN += len(set(range(0, n_points)) - set(true_labels[str(x)])- set(pred_labels[pred_labels['anoms'] == x]['ix']))
        
    CM = pd.DataFrame({"pred_T": {"label_T": TP, 
                                    "label_F": FP},
                         "pred_F": {"label_T": FN,
                                    "label_F": TN}}) 
    
    recall = CM['pred_T']['label_T'] / (CM['pred_T']['label_T'] + CM['pred_F']['label_T'])
    precision = CM['pred_T']['label_T'] / (CM['pred_T']['label_T'] + CM['pred_T']['label_F'])
    print('recall: ', recall)
    print('precision: ', precision)
    F1 = 2 * (precision * recall) / (precision + recall)
    return CM, F1


def gridsearch_MV_new(UBgrid, LBgrid, anoms, errors):
    grid = list(itertools.product(*[UBgrid, LBgrid]))
    perf = {}
    CM = {}
    TN = len(errors)-len(anoms)
    kernals = estimate_kernals(errors)
    for (UB,LB) in grid:
        anomalies = []
        for kernal in kernals:
            an = pd.DataFrame(anomaly_kernal_new(errors[kernal], kernals[kernal], UB, LB))
            an = an.rename(columns={0 : 'ix'})
            an['anoms'] = kernal
            anomalies.append(an)
        performance = F1_performance_new(anoms, pd.concat(anomalies), TN, len(errors.columns))
        perf[UB,LB] = performance[1]
        CM[performance[1]] = performance[0]
    
    best = max(perf, key=lambda key: perf[key])
    return CM[perf[best]], best, perf[best]
```

### Anomalies based on distribution distance (energy/wasserstein/ks_2samp)

```python
def plot_anomalies(series, anom_list):
    series.plot()
    for anom in anom_list:
        plt.scatter(x=series.iloc[anom:anom+1].index[0], y=series.iloc[anom], color='r')
    plt.show()
    

def cdf(data, steps):
    cdf=[]
    norm = sorted(data)
    for step in range(1, steps+1):
        cdf.extend([sum(norm[(int(len(data)/steps)*(step-1)):(int(len(data)/steps)*step-1)])/(int(len(data)/steps))])
    return cdf


def cdf_old(data, steps):
    cdf=[]
    norm = sorted(data)
    for step in range(1, steps+1):
        cdf.extend([norm[(int(len(data)/steps)*step-1)]])
    return cdf
```

```python
def error_distribution_distance(errors, timesteps):
    dfs = {}  #dict to store df's for timesteps
    for y in range(timesteps): 
        err = []
        for i in range(len(errors)):
            err.append(errors[i][y])
        dfs[y] = pd.DataFrame(err, index=range(y,(len(errors)+y))) #recenter predictions based on timestep index
    df_concat = pd.concat(dfs.values())
    
    total_kde = {}
    
    for column in pd.DataFrame(errors.reshape((len(errors)*timesteps), 6)):
        total_kde[column] = cdf(pd.DataFrame(errors.reshape((len(errors)*timesteps),6))[column], 24)

    dist_df = pd.DataFrame(columns=df_concat.columns, index=range(0, len(errors)))
    for label, df in df_concat.groupby(df_concat.index):
        if len(df) == 24:
            for column in df:
                dist_df.iloc[label][column] = energy_distance(total_kde[column], sorted(df[column]))
                #options: energy_distance, ks_2samp, wasserstein_distance
                #energy or wasserstein seem to perform similar, but ks suxxx

    return dist_df
```

```python
dis = error_distribution_distance(errors, 24)
```

```python
def gridsearch_MV_ED(UBgrid, anoms, errors):
    perf = {}
    CM = {}
    TN = len(errors)-len(anoms)
    for UB in UBgrid:
        anomalies = []
        for kernal in errors:
            an = pd.DataFrame(errors[errors[kernal] > UB].index)
            an = an.rename(columns={0 : 'ix'})
            an['anoms'] = kernal
            anomalies.append(an)
        performance = F1_performance_new(anoms, pd.concat(anomalies), TN, len(errors.columns))
        perf[UB] = performance[1]
        CM[performance[1]] = performance[0]
    best = max(perf, key=lambda key: perf[key])
    return CM[perf[best]], best, perf[best]
```

```python
def anomaly_kernal_MV_ED(errors, kernal, LB):
    anomalies = []
    for ix, point in enumerate(errors):
        integral = kernal.integrate_box_1d(point, np.inf)
        if integral < LB:
            anomalies.append(ix+23)
    return anomalies


def gridsearch_MV_ED_kern(UBgrid, anoms, errors):
    perf = {}
    CM = {}
    TN = len(errors)-len(anoms)
    kernals = estimate_kernals(errors)
    for UB in UBgrid:
        anomalies = []
        for kernal in errors:
            an = pd.DataFrame(anomaly_kernal_MV_ED(errors[kernal].values, kernals[kernal], UB))
            an = an.rename(columns={0 : 'ix'})
            an['anoms'] = kernal
            anomalies.append(an)
        performance = F1_performance_new(anoms, pd.concat(anomalies), TN, len(errors.columns))
        perf[UB] = performance[1]
        CM[performance[1]] = performance[0]
    best = max(perf, key=lambda key: perf[key])
    return CM[perf[best]], best, perf[best]
```
